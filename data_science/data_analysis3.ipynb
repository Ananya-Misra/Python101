```python
df.index
```

```python
geo_data= 'world_countries.json'
cols = [df.index,'total']
key ='feature.properties.name'
map = folium.Map()
map.choropleth(
    geo_data = geo_data,
    data = df,
    columns=cols,
    key_on = key,
    fill_color='YlOrRd',
    fill_opacity=.7,
    line_opacity=.3,
    legend_name='Immigration to CANADA from 1980-2013'
)

map
```
