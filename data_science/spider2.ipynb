```python
all_output=[]
query='laptop'
page=1
page_limit=10
while True:    
    url=f'https://www.flipkart.com/search?q={query}&page={page}'
    print(url)
    soup=sc.get_webpage_data(url)
    try:
        output=sc.extract_many(soup=soup,target=target,title=title,price=price,
        rr=rr,img_link=img_link,
        items=items,link=link,details=details)
        if output is not None:
            all_output.extend(output)
            page+=1
    except Exception as e:
        print(e)
        continue
    if page>page_limit:
        break
```
